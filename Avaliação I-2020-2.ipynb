### 1) Considerando duas listas de strings em que cada lista possua valores repetidos calcule o ângule vetorial através do cosseno, conforme equação abaixo:

O cosseno de dois vectores não zerados podem ser derivados utilizando a fórmula do produto dos pontos. A partir de dois vetores n-dimensionais, A e B, a similaridade do cosseno, cos(θ), é representado utilizando o produto dos pontos e a magnitude dos vetores como:


Obs: Será necessário importar a biblioteca math -> import math

```python
import math

list1 = ['w', 'w', 'e', 'f', 'j', 'h', 'i', 'j']
list2 = ['w', 'k', 'e', 'e', 'i', 'i', 'j', 'e']

dict1 = {}
for palavra in list1:
    if palavra not in dict1:
        dict1[palavra] = 0
    dict1[palavra] = dict1[palavra] + 1
print(dict1)
```

```python
dict2 = {}
for palavra in list2:
    if palavra not in dict2:
        dict2[palavra] = 0
    dict2[palavra] = dict2[palavra] + 1
print(dict2)
```

```python
num = 0
for palavra in dict1:
    if palavra in dict2:
        num += dict1[palavra]*dict2[palavra]
print(num)
```

```python
m1 = 0
for palavra in dict1:
    m1 += (dict1[palavra]*dict1[palavra])
print(m1)
    
```

```python
m2 = 0
for palavra in dict2:
    m2 += (dict2[palavra]*dict2[palavra])
print(m2)
```

```python
den = math.sqrt(m1) * math.sqrt(m2)
print(den)
```

```python
cos = num/den
print("COS(x) =", cos)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": [1, 2, 2, 3, 3, 4, 5, 6, 7]})
df
```

### 12) Considerando o conjunto de dados 2015_USPTOf.json (patentes americanas) realize as devidas transformações para que cada um dos valores que fazem parte do atributo "Subclass_labels", visto que o atributo é um <i>array</i>, se torne uma coluna que deve ser adicionada ao conjunto de dados original. Além disso, considerando determinada linha, as colunas que são identificadas pelos valores distintos do atributo "Subclass_labels" devem ter o valor 1 atribuido. Ou seja, somente a coluna que se refere a determinada classe de uma patente deve ser preenchida com 1. Ao final realize o somatório de cada uma das colunas adicionadas (referentes aos valores do atributo  "Subclass_labels") e apresente um histograma através de um gráfico de barras horizontal partindo das classes que mais ocorrem para àquelas que menos ocorrem. Apresente somente as 10 primeiras classes.

```python
import pandas as pd
import json

df = pd.read_json('2015_USPTOf.json')
```

```python
all_Subclass_labels = []
for x in df.Subclass_labels:
    all_Subclass_labels.extend(x)
Subclass_labels = pd.unique(all_Subclass_labels)
Subclass_labels
```

```python
zero_matrix = np.zeros((len(df), len(Subclass_labels)))
dummies = pd.DataFrame(zero_matrix, columns=Subclass_labels)
```

```python
for i, sub in enumerate(df.Subclass_labels):
    indices = dummies.columns.get_indexer(sub)
    dummies.iloc[i, indices] = 1
df_sub = df.join(dummies.add_prefix('Sub_label_'))
df_sub
```

```python
#prex_Subclass_labels = ['Sub_label_' + sub for sub in Subclass_labels]
dict_count = {}
for sub in Subclass_labels:
    dict_count[sub] = int(df_sub[str('Sub_label_' + sub )].sum())
#print(dict_count)
dict_count = sorted(dict_count.items(), key=lambda x: x[1], reverse=True)
print(dict_count)
```

```python
dict_count10 = dict(dict_count[0:10])
print(dict_count10)
```

```python
import matplotlib.pyplot as plt

keys= list(dict_count10.keys())
values = list(dict_count10.values())
print(keys)
print(values)
```

```python
plt.barh(keys,values)
plt.xlabel('Values')
plt.ylabel("Subclass_labels")
plt.title('Categories Bar Plot')
plt.show()
```
