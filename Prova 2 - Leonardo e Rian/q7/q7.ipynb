```python
# Importa a biblioteca de conjuntos de dados scikit-learn
import pandas as pd

# Carrega o conjunto de dados
nome_colunas = ['label', 'alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'nonflavanoid_phenols', 'proanthocyanins', 'color_intensity', 'hue', 'OD280/OD315_of_diluted_wines', 'proline']
# Carrega o dataset
# nao tem cabeçalho, logo header=None
# coloca o cabeçalho em nome_colunas
wine = pd.read_csv("wine.csv", header=None, names=nome_colunas)
wine['label'] = wine['label']-1
```

```python
wine.head()
```

```python
#Remove a coluna proline conforme solicitado
wine.drop('proline', axis=1, inplace=True)
```

```python
# Importa a função train_test_split
from sklearn.model_selection import train_test_split

X=wine[nome_colunas[1:13]]  # Catacterísticas (retirando o label)
y=wine['label']  # Rótulos

# Separa a conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% para treinamento e 30% para teste
```

```python
# Importa o classificador Random Forest
from sklearn.ensemble import RandomForestClassifier

# Cria o classificador
clf=RandomForestClassifier(n_estimators=100)

# Treina o modelo usando os conjuntos de treinamento (características e rótulo para predição)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
# Importa o módulo metrics de scikit-learn metrics para cálculo de acurácia
from sklearn import metrics

# Modelo de acurácia; com qual frequência o classificador é correto?
acc_origina = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", acc_origina)
```

```python
import pandas as pd
feature_imp = pd.Series(clf.feature_importances_,index=nome_colunas[1:13]).sort_values(ascending=False)
print("##########RELEVANCIA DE CADA CARACTERÍSTICA#############")
print(feature_imp)
```

```python
#Remove 6 das menos relevantes características 
new_wine = wine.drop(['nonflavanoid_phenols', 'proanthocyanins', 'ash', 'malic_acid', 'alcalinity_of_ash', 'magnesium'], axis=1, inplace=False)
new_wine
```

```python
#Refaz o treinamento com os novos dados sem 6 colunas

# Cria o classificador
clf_new=RandomForestClassifier(n_estimators=100)

# Treina o modelo usando os conjuntos de treinamento (características e rótulo para predição)
clf_new.fit(X_train,y_train)

y_pred_new=clf_new.predict(X_test)

#acuracia nova
acc_nova = metrics.accuracy_score(y_test, y_pred_new)
print("Accuracy:", acc_nova)
```

```python
print("Acuracia antes de remover 6 colunas menos importantes: ", acc_origina)
print("Acuracia depois de remover 6 colunas menos importantes: ", acc_nova)
```

```python
import numpy as np
from sklearn.cluster import KMeans

X = np.array(new_wine.drop(['label'], axis=1, inplace=False).astype(float))
y = np.array(new_wine['label'])

kmeans = KMeans(n_clusters=3) 
kmeans.fit(X)

correct = 0
for i in range(len(X)):
    predict_me = np.array(X[i].astype(float))
    predict_me = predict_me.reshape(-1, len(predict_me))
    prediction = kmeans.predict(predict_me)
    if (prediction[0]) == y[i]:
        correct += 1

print("Acuracia kmeans", correct/len(X))
```
